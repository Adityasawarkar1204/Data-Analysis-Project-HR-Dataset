# Data Analysis and Visulaization

### ANALYSIS ON CATEGORICAL COLUMN WRT TARGET COLUMN(ATTRITION)

```python
ax=sns.countplot(data=df, x='Attrition')
plt.show()
```

Observation:- Data of attrition say that it has more number of "No" values and less number of "Yes"

### 1. IMPACT OF BUSINESS TRAVEL ON ATTRITION

```python
ax=sns.countplot(data=df, x='Business Travel', hue='Attrition')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### 2. IMPACT OF DEPARTMENT ON ATTRITION

```python
ax2=sns.countplot(data=df, x='Department', hue='Attrition' )
plt.title('Distribution of Attrition by Department ')
for bars in ax2.containers:
    ax2.bar_label(bars)
plt.show()
```

### 3. IMPACT OF EDUCATION FIELD ON ATTRITION

```python
ax3=sns.countplot(data=df, x='Attrition', hue='Education Field')
for bars in ax3.containers:
    ax3.bar_label(bars)
plt.show()
```

### 4. IMPACT OF GENDER ON ATTRITION

```python
ax4=sns.countplot(data=df, x='Attrition', hue='Gender')
for bars in ax4.containers:
    ax4.bar_label(bars)
plt.show()
```

### 5. IMPACT OF OVERTIME ON ATTRITION

```python
ax5=sns.countplot(data=df,x='Over Time', hue='Attrition')
for bars in ax5.containers:
    ax5.bar_label(bars)
plt.show()
```

### 6. IMPACT OF JOB ROLE ON ATTRITION

```python
plt.figure(figsize=(15,6))
ax1=sns.countplot(data=df, x='Job Role', hue='Attrition')
plt.title('Distribution of Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.show()
```
